```python
import pandas as pd
import re
```

```python
all_trails = pd.read_pickle('all_trails_complete.pkl')
```

```python
all_trails.dropna(how = 'any', inplace=True)
```

```python
all_trails.shape
```

```python
all_trails = all_trails.reset_index(drop = True)
```

```python
all_trails.head()
```

```python
all_trails[['distance', 'elevation_gain', 'activities', 'photos', 'reviews', 'num_completed']] = all_trails[['distance', 'elevation_gain', 'activities','photos','reviews', 'num_completed']].replace(to_replace='[^\d\.]', value='', regex=True)
```

```python
all_trails[['distance', 'elevation_gain', 'activities', 'photos', 'reviews', 'num_completed']] = all_trails[['distance', 'elevation_gain', 'activities', 'photos', 'reviews','num_completed']].astype('float')
```

```python
all_trails.head()
```

```python
all_trails.dtypes
```

```python
all_trails['tags'] = all_trails['tags'].apply(', '.join)
```

```python
all_trails['dogs'] = all_trails.tags.str.contains('Dogs on leash').astype('int')
all_trails['views'] = all_trails.tags.str.contains('Views').astype('int')
all_trails['kids'] = all_trails.tags.str.contains('Kid friendly').astype('int')
all_trails['fee'] = all_trails.tags.str.contains('Fee').astype('int')
all_trails['wildlife'] = all_trails.tags.str.contains('Wildlife').astype('int')
all_trails['lake'] = all_trails.tags.str.contains('Lake').astype('int')
all_trails['beach'] = all_trails.tags.str.contains('Beach').astype('int')
all_trails['wildflowers'] = all_trails.tags.str.contains('Wildflowers').astype('int')
all_trails['rocky'] = all_trails.tags.str.contains('Rocky').astype('int')
all_trails['loop'] = all_trails.route_type == 'Loop'
all_trails.loop = all_trails.loop.astype('int')
all_trails = pd.get_dummies(all_trails, columns = ['difficulty'], drop_first = True)
```

```python
all_trails.head()
```

```python
all_trails.drop(columns = ['route_type','tags'], inplace = True)
```

```python
all_trails.head()
```

```python
all_trails.to_pickle('cleaned_trails.pkl')
```
